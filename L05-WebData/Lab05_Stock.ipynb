```python
df=pd.DataFrame(data['AAPL'],index=data.index,columns=['AAPL'])
df.tail()
```

```python
df['Return']=0.0
for i in range(1,len(df)):
    df['Return'][i]=np.log(df['AAPL'][i]/df['AAPL'][i-1])
df.tail()
```

```python
df['Return2']=np.log(df['AAPL']/df['AAPL'].shift(1))
df[['AAPL','Return','Return2']].tail()df
```

```python
del df['Return2']
df.tail()
```

```python
df[['AAPL','Return']].plot(subplots=True,style=['b','r'],figsize=(8,5))
```

```python
df['42d']=pd.rolling_mean(df['AAPL'],window=42)
df['242d']=pd.rolling_mean(df['AAPL'],window=242)
df[['AAPL','42d','242d']].tail()
```

```python
df[['AAPL','42d','242d']].head()
```

```python
df[['AAPL','42d','242d']].plot(figsize=(8,5))
```

```python
import math
df['Mov_Vol']=pd.rolling_mean(df['AAPL'],window=252)*math.sqrt(252)
df['Mov_Vol'].tail()
```

```python
df[['AAPL','Mov_Vol','Return']].plot(subplots=True,style=['b','g','r'],figsize=(8,8))
```

```python
df[['AAPL','Mov_Vol','Return']].plot(figsize=(8,5))
```
